```python
app_details_csv = pd.read_csv('playstorescrapy/spiders/final_top_apps_details.csv')
app_details_csv.shape
```

```python
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns',90)
```

```python
from numpy import nan

def detect_lang(text): # return language code
    try:
        DetectorFactory.seed=0
        return detect(text)
    except:
        return nan

def is_english(text): # returns Boolean
    text= text.lower()
    # checking for ascii value of characters
    for word in text.split(' '):
        if ord(word[0]) >= 97 and ord(word[0])<=122:
            return True
    return False

def find_country(address): 
    if not address :
        return nan

    if is_english(address):  
        for country in all_countries:
            index = address.lower().find(country.lower())
            if index>-1:
                return country
        return nan
    else:
        lang_code = detect_lang(address)
        return lang_map[lang_code] if lang_code in lang_map else lang_code
```

```python
app_details_csv['developer_address'] = app_details_csv['developer_address'].astype(str)
app_details_csv['app_country'] = app_details_csv['developer_address'].apply(find_country)
```
